Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# printing the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
# number of rows and Columns in this dataset
diabetes_dataset.shape
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

0 --> Non-Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

Making a Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = classifier.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```

```python
# =========================================
#  AI Health Dashboard (Futuristic Glass + New Gradient Palette + Stylish Greeting)
# =========================================

!pip install gradio==4.44.0 matplotlib scikit-learn joblib --quiet

import os, joblib, numpy as np, matplotlib.pyplot as plt, gradio as gr
from pathlib import Path

MODEL_DIR = Path("/content/saved_models")

# -----------------------------
# Safe model loader
# -----------------------------
def load_model_checked(path):
    info = {"model": None, "error": None, "n_features": None}
    if not Path(path).exists():
        info["error"] = f"File not found: {path}"
        return info
    try:
        m = joblib.load(path)
        info["model"] = m
        n = getattr(m, "n_features_in_", None)
        if n is not None:
            info["n_features"] = int(n)
        return info
    except Exception as e:
        info["error"] = f"Failed to load: {e}"
        return info

model_paths = {
    "diabetes": MODEL_DIR / "diabetes_model.sav",
    "heart": MODEL_DIR / "heart_disease_model.sav",
    "parkinsons": MODEL_DIR / "parkinsons_model.sav",
}
models_info = {k: load_model_checked(str(p)) for k, p in model_paths.items()}

# -----------------------------
# Users + history
# -----------------------------
USERS, history = {"test": "123"}, {}

# -----------------------------
# Prediction utils
# -----------------------------
def validate_and_prepare(inputs, disease):
    try:
        arr = np.array(inputs, dtype=float)
    except Exception as e:
        return False, f"Invalid inputs: {e}"
    if np.isnan(arr).any():
        return False, " Inputs contain NaN values."
    if np.all(arr == 0):
        return False, " Please enter non-zero values."

    info = models_info[disease]
    if info["model"] is None:
        return False, f"Model for {disease} not loaded."

    if info["n_features"] is not None and arr.size != info["n_features"]:
        return False, f"Expected {info['n_features']} features, got {arr.size}."

    return True, arr.reshape(1, -1)

def predict(disease, inputs, user):
    ok, res = validate_and_prepare(inputs, disease)
    if not ok:
        return f"<div style='color:red;font-weight:bold'>{res}</div>"

    X = res
    model_info = models_info[disease]
    model = model_info["model"]

    if model is None:
        return f"<div style='color:red;font-weight:bold'>Model for {disease} not loaded!</div>"

    try:
        pred = model.predict(X)
        pred0 = int(pred[0])
    except Exception as e:
        return f" Prediction failed: {e}"

    label_map = {
        "diabetes": {0:" No Diabetes", 1:" Diabetes Detected"},
        "heart": {0:" No Heart Disease", 1:" Heart Disease Detected"},
        "parkinsons": {0:" No Parkinsons", 1:" Parkinsons Detected"}
    }
    label = label_map[disease].get(pred0, " Unknown prediction")
    color = "rgba(100,247,141,0.3)" if "No" in label else "rgba(234,34,100,0.3)"  # Green for No, Red for Detection

    history.setdefault(user, []).append((disease, label))

    return f"""
    <div style='
        background:{color};
        backdrop-filter: blur(12px);
        -webkit-backdrop-filter: blur(12px);
        border-radius:20px;
        color:#fff;
        padding:15px;
        font-weight:bold;
        font-size:18px;
        text-align:center;
        box-shadow:0 8px 25px rgba(0,0,0,0.1);
        transition:0.3s;
    '>{label}</div>
    """

# -----------------------------
# Auth + history
# -----------------------------
def signup(u, p):
    if not u or not p:
        return " Provide username & password!"
    if u in USERS:
        return " User exists!"
    USERS[u] = p
    return " Signup successful!"

def login(u, p):
    if USERS.get(u) == p:
        greet_html = f"""
        <div class="greet-card">
            <span class="wave"></span> Hello <b>{u}</b>!
        </div>
        """
        return (
            f" Welcome {u}!",
            u,
            gr.update(value=greet_html, visible=True),
            gr.update(visible=False),
            gr.update(visible=True),
        )
    return (
        " Invalid credentials!",
        "",
        gr.update(value="", visible=False),
        gr.update(visible=True),
        gr.update(visible=False),
    )

def logout():
    return (
        "",
        "",
        gr.update(value="", visible=False),
        gr.update(visible=True),
        gr.update(visible=False),
    )

def show_history(u):
    if not history.get(u):
        return " No history!"
    return "\n".join([f" {d.title()}: {r}" for d,r in history[u]])

def show_pie(u):
    if not history.get(u): return None
    counts = {}
    for _, r in history[u]:
        counts[r] = counts.get(r, 0) + 1
    fig, ax = plt.subplots(figsize=(5,5))
    labels, sizes = list(counts.keys()), list(counts.values())
    colors = ["#64F78D" if "No" in k else "#EA2264" for k in labels]  # Green/Red
    wedges, _, autotexts = ax.pie(
        sizes, autopct='%1.1f%%', startangle=140, colors=colors,
        pctdistance=0.8, wedgeprops=dict(width=0.4, edgecolor='white', linewidth=2)
    )
    for at in autotexts:
        at.set_color("#fff")
        at.set_weight("bold")
    ax.set_title("Prediction Outcomes", fontsize=14, weight="bold")
    ax.legend(wedges, labels, title="Results", loc="center left", bbox_to_anchor=(1,0,0.5,1))
    plt.tight_layout()
    return fig

# -----------------------------
# Input ranges + default values
# -----------------------------
# Diabetes
diabetes_ranges = [(0,17),(0,199),(0,122),(0,99),(0,846),(0,67.1),(0.078,2.42),(21,81)]
diabetes_defaults = [6,148,72,35,0,33.6,0.627,50]

# Heart
heart_ranges = [(29,77),(0,1),(0,3),(94,200),(126,564),(0,1),(0,2),(71,202),(0,1),(0,6.2),(0,2),(0,4),(0,3)]
heart_defaults = [68,1,0,144,193,1,1,141,0,3.4,1,2,3]

# Parkinson's
parkinsons_ranges = [
    (88.33,260.11),(102.15,592.03),(65.48,239.17),(0.00168,0.03316),(0.000007,0.00026),
    (0.00068,0.02144),(0.00092,0.01958),(0.00204,0.06433),(0.00954,0.11908),(0.085,1.302),
    (0.00455,0.05647),(0.0057,0.0794),(0.00719,0.13778),(0.01364,0.16942),(0.00065,0.31482),
    (8.441,33.047),(0.25657,0.68515),(0.57428,0.82529),(-7.96498,-2.43403),(0.00627,0.45049),
    (1.42329,3.67116),(0.04454,0.52737)
]
parkinsons_defaults = [
    119.99200, 157.30200, 74.99700, 0.00784, 0.00007, 0.00370, 0.00554, 0.01109, 0.04374, 0.42600,
    0.02182, 0.03130, 0.02971, 0.06545, 0.02211, 21.03300, 0.414783, 0.815285, -4.813031, 0.266482,
    2.301442, 0.284654
]

# -----------------------------
# UI Layout
# -----------------------------
with gr.Blocks(theme="default") as demo:
    state = gr.State("")

    # Login Panel
    with gr.Group(visible=True) as login_panel:
        gr.Markdown("<h1 style='text-align:center; color:#EA2264'> AI Health Dashboard</h1>")
        u = gr.Textbox(label=" Username")
        p = gr.Textbox(label=" Password", type="password")
        with gr.Row():
            login_btn = gr.Button("Login ", elem_classes="btn-purple")
            signup_btn = gr.Button("Signup ", elem_classes="btn-orange")
        auth_msg = gr.Markdown("")

    # Dashboard
    with gr.Group(visible=False) as dash:
        with gr.Row():
            # Sidebar
            with gr.Column(scale=1, elem_classes="glass-panel"):
                dash_greet = gr.HTML("", elem_classes="greet-card", visible=False)
                diabetes_nav = gr.Button(" Diabetes", elem_classes="nav-btn")
                heart_nav = gr.Button(" Heart", elem_classes="nav-btn")
                parkinsons_nav = gr.Button(" Parkinsons", elem_classes="nav-btn")
                history_nav = gr.Button(" History & Analytics", elem_classes="nav-btn")
                logout_btn = gr.Button(" Logout", elem_id="logout-btn")
            # Main content
            with gr.Column(scale=4):
                # Diabetes
                with gr.Group(visible=True, elem_classes="glass-panel") as diabetes_panel:
                    gr.Markdown("###  Diabetes Prediction", elem_classes="card-title")
                    di_labels = ["Pregnancies","Glucose","BP","Skin","Insulin","BMI","DPF","Age"]
                    di_in = [gr.Number(label=l, value=v, minimum=r[0], maximum=r[1], step=0.01)
                             for l,v,r in zip(di_labels, diabetes_defaults, diabetes_ranges)]
                    di_btn = gr.Button(" Predict Diabetes", elem_classes="btn-purple")
                    di_out = gr.HTML()

                # Heart
                with gr.Group(visible=False, elem_classes="glass-panel") as heart_panel:
                    gr.Markdown("###  Heart Disease Prediction", elem_classes="card-title")
                    he_labels = ["Age","Sex","CP","Trestbps","Chol","Fbs","Restecg","Thalach","Exang","Oldpeak","Slope","Ca","Thal"]
                    he_in = [gr.Number(label=l, value=v, minimum=r[0], maximum=r[1], step=0.01)
                             for l,v,r in zip(he_labels, heart_defaults, heart_ranges)]
                    he_btn = gr.Button(" Predict Heart Disease", elem_classes="btn-purple")
                    he_out = gr.HTML()

                # Parkinson's
                with gr.Group(visible=False, elem_classes="glass-panel") as parkinsons_panel:
                    gr.Markdown("###  Parkinsons Prediction", elem_classes="card-title")
                    pa_labels = [
                        "Fo","Fhi","Flo","Jitter(%)","Jitter(Abs)","RAP","PPQ","DDP",
                        "Shimmer","Shimmer(dB)","APQ3","APQ5","APQ","DDA","NHR","HNR",
                        "RPDE","DFA","spread1","spread2","D2","PPE"
                    ]
                    pa_in = [gr.Number(label=l, value=v, minimum=r[0], maximum=r[1], step=0.0001 if i>=3 and i<=15 else 0.01)
                             for i,(l,v,r) in enumerate(zip(pa_labels, parkinsons_defaults, parkinsons_ranges))]
                    pa_btn = gr.Button(" Predict Parkinsons", elem_classes="btn-purple")
                    pa_out = gr.HTML()

                # History
                with gr.Group(visible=False, elem_classes="glass-panel") as history_panel:
                    gr.Markdown("###  Your History & Analytics", elem_classes="card-title")
                    his_btn = gr.Button(" Show History", elem_classes="btn-orange")
                    his_out = gr.Textbox(label="History", lines=10, interactive=False, elem_classes="history-card")
                    pie_btn = gr.Button(" Show Analytics", elem_classes="btn-orange")
                    pie_out = gr.Plot()

    # Wiring
    signup_btn.click(signup, [u,p], [auth_msg])
    login_btn.click(login, [u,p], [auth_msg, state, dash_greet, login_panel, dash])
    logout_btn.click(logout, None, [auth_msg, state, dash_greet, login_panel, dash])

    di_btn.click(lambda *x: predict("diabetes", x[:-1], x[-1]), di_in + [state], di_out)
    he_btn.click(lambda *x: predict("heart", x[:-1], x[-1]), he_in + [state], he_out)
    pa_btn.click(lambda *x: predict("parkinsons", x[:-1], x[-1]), pa_in + [state], pa_out)

    his_btn.click(show_history, [state], [his_out])
    pie_btn.click(show_pie, [state], [pie_out])

    def nav_switch(active):
        return (
            gr.update(visible=(active=="diabetes")),
            gr.update(visible=(active=="heart")),
            gr.update(visible=(active=="parkinsons")),
            gr.update(visible=(active=="history")),
        )
    diabetes_nav.click(lambda: nav_switch("diabetes"), None, [diabetes_panel,heart_panel,parkinsons_panel,history_panel])
    heart_nav.click(lambda: nav_switch("heart"), None, [diabetes_panel,heart_panel,parkinsons_panel,history_panel])
    parkinsons_nav.click(lambda: nav_switch("parkinsons"), None, [diabetes_panel,heart_panel,parkinsons_panel,history_panel])
    history_nav.click(lambda: nav_switch("history"), None, [diabetes_panel,heart_panel,parkinsons_panel,history_panel])

# -----------------------------
# CSS with new palette + Greeting Animation
# -----------------------------
demo.css = f"""
body {{
    background: linear-gradient(135deg, #0D1164, #640D5F, #EA2264, #F78D60);
    min-height:100vh;
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    overflow-x:hidden;
    position: relative;
}}
body::before, body::after {{
    content:''; position:absolute; width:600px; height:600px;
    background: radial-gradient(circle, rgba(100,247,141,0.4), rgba(15,15,15,0));
    border-radius:50%;
    animation: float1 20s ease-in-out infinite alternate;
    z-index:-1;
}}
body::after {{
    top:50%; left:60%;
    background: radial-gradient(circle, rgba(234,34,100,0.4), rgba(247,141,96,0));
    animation: float2 25s ease-in-out infinite alternate;
}}
@keyframes float1 {{0%{{transform:translate(0,0)}}50%{{transform:translate(100px,-50px)}}100%{{transform:translate(0,-100px)}}}}
@keyframes float2 {{0%{{transform:translate(0,0)}}50%{{transform:translate(-100px,50px)}}100%{{transform:translate(0,100px)}}}}

/* Greeting Card */
.greet-card {{
    background: linear-gradient(135deg, rgba(100,13,95,0.7), rgba(234,34,100,0.7));
    backdrop-filter: blur(12px);
    -webkit-backdrop-filter: blur(12px);
    border-radius: 16px;
    padding: 14px 20px;
    text-align: center;
    font-size: 18px;
    font-weight: bold;
    color: #fff;
    box-shadow: 0 6px 18px rgba(0,0,0,0.25);
    transition: transform 0.3s ease, box-shadow 0.3s ease;
    margin-bottom: 12px;
}}
.greet-card:hover {{
    transform: translateY(-3px);
    box-shadow: 0 10px 25px rgba(0,0,0,0.35);
}}
.wave {{
    display:inline-block;
    animation: wave 1.5s infinite;
}}
@keyframes wave {{
    0% {{ transform: rotate(0deg); }}
    20% {{ transform: rotate(20deg); }}
    40% {{ transform: rotate(-10deg); }}
    60% {{ transform: rotate(20deg); }}
    80% {{ transform: rotate(-5deg); }}
    100% {{ transform: rotate(0deg); }}
}}

/* Glass panels */
.glass-panel {{background: rgba(255,255,255,0.15); backdrop-filter: blur(14px); -webkit-backdrop-filter: blur(14px); border-radius:20px; padding:15px; box-shadow:0 8px 30px rgba(0,0,0,0.1); transition: transform 0.3s, box-shadow 0.3s;}}
.glass-panel:hover {{transform: translateY(-3px); box-shadow: 0 12px 35px rgba(0,0,0,0.2);}}

/* Buttons */
.nav-btn{{background:linear-gradient(135deg,#0D1164,#640D5F);color:#fff;border-radius:12px;margin:6px 0;padding:10px;font-weight:bold;transition:0.3s;}}
.nav-btn:hover{{transform:translateX(5px);background:linear-gradient(135deg,#EA2264,#F78D60);}}
.btn-purple{{background:linear-gradient(135deg,#640D5F,#EA2264);color:#fff;border-radius:12px;font-weight:bold;padding:10px 15px;}}
.btn-purple:hover{{background:linear-gradient(135deg,#F78D60,#EA2264);}}
.btn-orange{{background:linear-gradient(135deg,#F78D60,#EA2264);color:#fff;border-radius:12px;font-weight:bold;padding:10px 15px;}}
.btn-orange:hover{{background:linear-gradient(135deg,#EA2264,#F78D60);}}
#logout-btn{{background:linear-gradient(135deg,#EA2264,#640D5F);color:#fff;border-radius:12px;margin-top:20px;font-weight:bold;}}
#logout-btn:hover{{background:linear-gradient(135deg,#640D5F,#EA2264);}}

/* Titles */
.card-title{{color:#fff;font-size:20px;font-weight:bold;text-align:center;margin-bottom:12px;}}

/* History */
.history-card{{background:rgba(255,255,255,0.1);border-radius:12px;color:#fff;padding:12px;font-weight:bold;}}
"""

demo.launch(share=True)
```
